```python
import jax.numpy as jnp
import jax
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = 5 #number of holidays
t = jax.numpy.expand_dims(jnp.arange(50), 1) #days
holidays_matrix = jnp.zeros((len(t),d))
key = jax.random.PRNGKey(777)
kappa = jax.random.normal(key, (d,))
```

```python
np.random.seed(420)
for i in range(d-1):
    indexes = np.random.choice(np.arange(50), 3)
    holidays_matrix = holidays_matrix.at[indexes,i].set(1)
```

```python
indexes = np.arange(0,50,7,dtype=int)
holidays_matrix = holidays_matrix.at[indexes,4].set(1)
```

```python
holidays = jax.numpy.sum(holidays_matrix * kappa, axis=1)
```

```python
plt.plot(jnp.arange(50),holidays)
plt.show()
```
